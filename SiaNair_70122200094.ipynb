```python
import torch
from diffusers import DiffusionPipeline
import os
import time
```

* I used Stable Diffusion XL (SDXL Base), which is an open-source text-to-image model. It gives better quality and realism than the older Stable Diffusion versions. Industry-standard open-source diffusion model with 1024×1024 native resolution.

* Everything was run using Hugging Face’s diffusers library, which makes it simple to load and generate images from these models.

* The code was optimized for GPU (CUDA) with torch_dtype=torch.float16 so that it runs faster and uses less memory.

```python
MODEL_ID = "stabilityai/stable-diffusion-xl-base-1.0"
# Base directory for all generated images
BASE_OUTPUT_DIR = "assignment_images_organized"
os.makedirs(BASE_OUTPUT_DIR, exist_ok=True)
```

```python
print(f"Loading open-source model: {MODEL_ID}...")
device = "cuda" if torch.cuda.is_available() else "cpu"
start_time = time.time()
```

```python
pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)
```

```python
pipe.to("cuda")
print(f"Model loaded in {time.time() - start_time:.2f} seconds.")
```

```python
prompts = [
    # Location 1: Mumbai Bandra-Worli Sea Link
    {
        "name": "Elevated_Blue_Hour_Car",
        "location_folder": "Mumbai_Sea_Link",
        "prompt": "A breathtaking, cinematic wide-angle photograph of a **silver Mahindra XUV 700 SUV** parked at an elevated viewpoint. The **Mumbai Bandra-Worli Sea Link** must be in the background, spanning the entire frame with **perfect structural integrity** and fully illuminated at blue hour. The car's **isolated silver color** is prominent. Moody blue sky, sharp focus on both the car and the landmark. Professional automotive photography, 16k, high-dynamic range, **clear view of the landmark, no structural collapse.**"
    },
    {
        "name": "Low_Angle_Pylon_Car",
        "location_folder": "Mumbai_Sea_Link",
        "prompt": "A dramatic, vertical low-angle shot of a **silver Mahindra XUV 700 SUV** driving on the road. The background is the **Bandra-Worli Sea Link**'s massive concrete pylon and crisscrossing steel cables, which must be structurally accurate. The car is sharp, detailed, and the **silver color** is not tinted. Slight motion blur on the wheels, with a moody monsoon sky overhead. Hyper-detailed, architectural and automotive photography, **perfect structural integrity, clear bridge.**"
    },
    # Location 2: Rajasthan Hawa Mahal
    {
        "name": "Frontal_Street_Car",
        "location_folder": "Hawa_Mahal",
        "prompt": "The classic frontal facade of the **Hawa Mahal (Palace of Winds)** in Jaipur, Rajasthan, serving as the background. In the foreground, a **silver Mahindra XUV 700 SUV** is parked on the street. **The car must be rendered in pure polished metallic silver.** The ornate, pink-sandstone Mahal is in sharp focus, and the car's **silver color is isolated from ambient pink reflections.** Street photography, vibrant colors, sunny day, quality background, cinematic realism, **NO PINK CAR.**"
    },
    {
        "name": "Side_Angle_SUV_Car",
        "location_folder": "Hawa_Mahal",
        "prompt": "A side-angle photograph of the **Hawa Mahal**, showcasing its intricate side windows and pink walls as the background. In the foreground, a **silver Mahindra XUV 700 SUV** is parked. The car's **silver metallic paint** must be prominently displayed. Golden hour lighting, capturing the texture of the sandstone and the gloss of the car. Professional automotive advertising photo, **ultra realistic photograph, car isolated silver, NO PINK HUE on car.**"
    },
    # Location 3: Nagpur rainforest
    {
        "name": "Forest_Jeep_on_Track",
        "location_folder": "Nagpur_Rainforest",
        "prompt": "A photorealistic image of a rugged, **silver Mahindra XUV 700 SUV** driving down a narrow dirt track in a dense, central Indian jungle (**Nagpur rainforest**). The SUV is dusty and prominent. The background features tall trees, dense foliage, and sunlight dappling through the canopy. Wildlife photography style, deep perspective, adventure scene, **photorealistic quality.**"
    },
    {
        "name": "Misty_Luxury_SUV_Car",
        "location_folder": "Nagpur_Rainforest",
        "prompt": "A vertical close-up shot of the front of a **silver Mahindra XUV 700 SUV** parked deep within a lush, moist forest (**Nagpur rainforest**). The SUV is framed by dense, vibrant green ferns and moss-covered rocks. A thick mist or fog hangs around the scene. The focus is on the car and the rich, humid background foliage. Atmospheric, macro-style nature and automotive photo, **photorealistic quality.**"
    },
]
```

```python
negative_prompt = "lowres, worst quality, low quality, glitch, bad anatomy, deformed, blurry, jpeg artifacts, monochrome, multiple cars, distant or tiny car, incorrect car model, **pink car**, **pink tint on car**, washed out colors, structural errors, deformed structure, collapsed bridge, amateur photo"

for i, task in enumerate(prompts):
    location_dir = os.path.join(BASE_OUTPUT_DIR, task["location_folder"])
    os.makedirs(location_dir, exist_ok=True)

    print(f"Generating image {i + 1}/6 for {task['location_folder']}...")

    # Generate the image
    image = pipe(
        prompt=task["prompt"],
        negative_prompt=negative_prompt,
        num_inference_steps=30,
        guidance_scale=8.5,
        width=1024,
        height=1024
    ).images[0]

    # Save the generated image to the new location-specific folder
    file_path = os.path.join(location_dir, f"{task['name']}.png")
    image.save(file_path)
    print(f"Saved to: {file_path}")
```

```python
print(f"The six unique images have been successfully saved into three separate folders within the '{BASE_OUTPUT_DIR}' directory.")
```

# Steps I Followed

* Step 1: Setup

Loaded the SDXL model and made sure outputs get saved into separate folders for each location (Mumbai_Sea_Link, Hawa_Mahal, Nagpur_Rainforest).

* Step 2: Prompt Engineering

Wrote detailed prompts to describe the car, the scene, and the style.

Special care was taken for:

Car color: making sure it always stayed silver (no pink tint from Hawa Mahal background).

Landmarks: bridge structure, palace details, and rainforest trees looked accurate.

Lighting and mood: used terms like blue hour, golden hour, misty rainforest for realism.

Added a negative prompt to filter out issues like: wrong car model, pink car, blurry/low-quality outputs, or broken structures.

* Step 3: Generation

For each of the six cases, generated the image with:

num_inference_steps=30  enough for quality while keeping it fast.

guidance_scale=8.5  ensures the output follows the text prompt closely.

Image size set to 1024x1024 for sharpness.

Saved all images neatly into their respective folders so the results are organized and easy to review.
